```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Bengaluru_House_Data.csv')
```

```python
data.head()
```

```python
data
```

```python
data.drop(columns=['area_type', 'society', 'availability'], inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data
```

now data cleaning is done

```python
data.location.value_counts()
```

there are 1 house in some location, like rare areas. this is bad to feed to the model. we should somewhat balance the data

```python
data['location'] = data['location'].apply(lambda x:x.strip())
```

```python
data.location.value_counts()
```

```python
location_stats = data.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
location_less_than_10_entries=location_stats[location_stats <= 10]
```

```python
location_less_than_10_entries
```

```python
data['location'] = data['location'].apply(lambda x: 'other' if x in location_less_than_10_entries.index else x)
```

```python
data['location'].value_counts()
```

```python
data
```

```python
data['size'].unique()
```

here we can see that we have bhk, bedroom, Rk and so on. we shall now spilt them and get the numeric values only

```python
data['bedrooms'] = data['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
data
```

```python
data['total_sqft'].unique()
```

we can see there are values in form of ranges, we will clean that too

```python
def clean_sqft(x):
    try:
        tokens = x.split('-')
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1])) / 2
        return float(x)
    except:
        return None
```

```python
clean_sqft('2100 - 2850')
```

```python
clean_sqft('hello')
```

```python
data['total_sqft'] = data['total_sqft'].apply(clean_sqft)
```

```python
data['total_sqft'].unique()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.describe()
```

```python
data['sqft_per_bedroom'] = data['total_sqft'] / data['bedrooms']
```

```python
data
```

```python
data.sqft_per_bedroom.describe()
```

min for sqft per bedroom is 0.71 which is like a invalid. so we will apply condition where min sqft per bedroom should more or equal to 300sqft

```python
data2 = data[data['sqft_per_bedroom'] >= 300]
```

```python
data2
```

```python
data2['price_per_sqft'] = round(data2['price']*100000 / data2['total_sqft'],2)
```

```python
data2
```

```python
data2.price_per_sqft.describe()
```

again here min is 267, which is kind of a invalid. lets consider atleast 2000Rs so atleast it is judgable

```python
data3 = data2[data2['price_per_sqft'] >= 2000]
```

```python
data3
```

```python
data3.drop(columns=['size', 'sqft_per_bedroom', 'price_per_sqft'], axis=1,inplace=True)
```

```python
data3
```

we have location values as categorical value and we have to convert it to numerical.

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
col_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']),remainder='passthrough')
```

```python
lr = LinearRegression()
scaler = StandardScaler()
```

```python
model = make_pipeline(col_trans, scaler, lr)
```

```python
data_input = data3.drop(columns=['price'], axis=1)
data_output = data3['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_input, data_output, test_size=0.2)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
input = pd.DataFrame([['Electronic City Phase II', 2000.0, 3.0, 2.0, 3]], columns=['location', 'total_sqft', 'bath','balcony', 'bedrooms'])
```

```python
model.predict(input)
```

```python
import pickle
pickle.dump(model, open('banglore_home_price_model.pkl', 'wb'))
```
